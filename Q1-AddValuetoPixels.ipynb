# Q1. program to add value to all pixel of an image & display the image


```python
from PIL import Image
```

```python
img = Image.open("img1.jpg")
print("Original Image")
display(img)
```

```python
val = int(input("Enter a Value:= "))
```

```python
for y in range(img.height):
    for x in range(img.width):
        pixel = img.getpixel((x,y))
        red_pixel = pixel[0]+val
        green_pixel = pixel[1]+val
        blue_pixel = pixel[2]+val
        img.putpixel((x,y),(red_pixel,green_pixel,blue_pixel))

img.save("q1_output.jpg")
print("image after operation")
display(img)
```
