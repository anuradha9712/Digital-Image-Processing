# Q13 Program to perform Histogram Equalisation of an Image


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=Image.open("img1.jpg").convert("L")
```

```python
print(img.height)
print(img.width)
```

```python
arr1=np.empty([405,720]) #to store intensities in matrix
arr2=np.empty(256)       #store frequencies
for x in range (0,256):
    arr2[x]=0
```

```python
def histogram(arr1):
    for y in range(0,197):
        for x in range(0,255):
            i=int(arr1[y][x])
            arr2[i]=arr2[i]+1
```

```python

for y in range(img.height):
    for x in range(img.width):
        pixels=img.getpixel((x,y))
        arr1[y][x]=pixels

histogram(arr1)
```

```python
#probability calculation
arr3=np.empty(256)
for x in range(0,256):
    arr3[x]=arr2[x]/(405*720)

for x in range(0,256):
    print(arr3[x])
```

```python
#transformation fn or cdf
arr4=np.empty(256)
for x in range(0,256):
    if (x==0):
        arr4[x]=256*arr3[x]
    else:
        arr4[x]=(256*arr3[x])+arr4[x-1]  
        
#print output values
for x in range(0,256):
    print(arr4[x])
```

```python
#roundoff values
arr5=np.empty(256)
for x in range(0,256):
    arr5[x]=int(arr4[x])
    arr5[x].astype(int)    
#print output values
for x in range(0,256):
    print(arr5[x])
```

```python
#s values by adding r values
arr6=np.empty(256)
i=0.0
j=0
for x in range(0,256):
    if(arr5[x]==i):
        arr6[j]+=arr3[x]
    else:
        j+=1
        i+=1.0
        
for x in range(0,j+1):
    print(arr6[x])
```

```python
arr7=np.empty(256)
for x in range(0,256):# to plot frequency
       arr7[x]=x
plt.bar(arr7,arr3)
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Image")
plt.show()
```

```python
plt.bar(arr7,arr6)
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.title("Histogram of New Image")
plt.show()
```
