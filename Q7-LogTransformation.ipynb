# Log Transformation of gray image

```python
#Log Transformation
#resultant_pixel = c * log(actual_pixel + 1)
#where c is a constant calculated as 
#c = 255/log(maximum_pixel_value + 1)

from PIL import Image
import math

img=Image.open("img1.jpg")


def log_Transform(r):
    c=255/math.log((1+255),10) #use log base 10
    s=c*math.log((1+r),10)
    return s

for y in range(img.height):
    for x in range(img.width):
        current_img=img.getpixel((x,y))
        red_pixel=int(log_Transform(current_img[0]))
        green_pixel=int(log_Transform(current_img[1]))
        blue_pixel=int(log_Transform(current_img[2]))
        img.putpixel((x,y),(red_pixel,green_pixel,blue_pixel))

#img.show()
display(img)
img.save("q7_output.jpg")
```
