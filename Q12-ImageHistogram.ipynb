# Histogram of an Image

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img=Image.open("img1.jpg").convert("L")
display(img)
```

```python
print(img.height)
print(img.width)
```

```python
arr1=np.empty([405,720])  #to store intensities in matrix
arr2=np.empty(256)        #store frequencies
for x in range (0,256):
    arr2[x]=0
```

```python
def histogram(arr1):
    for y in range(0,405):
        for x in range(0,720):
            i=int(arr1[y][x])
            arr2[i]=arr2[i]+1
```

```python
for y in range(img.height):
    for x in range(img.width):
        pixels=img.getpixel((x,y))
        arr1[y][x]=pixels

histogram(arr1)
```

```python
arr3=np.empty(256)
for x in range(0,256):# to plot frequency
       arr3[x]=x
plt.bar(arr3,arr2)
plt.title("HISTOGRAM OF AN IMAGE")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.show()
```
