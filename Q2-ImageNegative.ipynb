# Q2. To Compute Negative of an image

```python
from PIL import Image
```

```python
img = Image.open("img1.jpg")
print("original image")
display(img)
```

```python
for y in range(img.height):
    for x in range(img.width):
        pixel = img.getpixel((x,y))
        red_pixel = 255-pixel[0]
        green_pixel = 255-pixel[1]
        blue_pixel = 255-pixel[2]
        img.putpixel((x,y),(red_pixel,green_pixel,blue_pixel))
        
#img.show()
print("Negative of image")
display(img)
img.save("q2_output.jpg")
```
